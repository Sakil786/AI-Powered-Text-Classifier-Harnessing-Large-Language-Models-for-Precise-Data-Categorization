# **Creating a function for converting the data in Alpaca format**



*   The dataset is divided into inference data and training data
*   First 1000 datapoints is taken as inference data and remaining data point is considered for training data



```python
def get_training_data_preparation():
  # Function to prepare training data
  training_data=pd.read_csv('/content/train_40k.csv')
  training_data.dropna(inplace=True)
  inference_data=training_data.iloc[:1000]
  final_train_data=training_data.iloc[1001:]
  # final_train_data['Cat'] = final_train_data['Cat2'] + ',' + final_train_data['Cat3']
  # converting the data in alpaca format
  final_train_data['instruction']='Classify the text into categories given in output column. Reply with only the words given in output column.'
  # For datapreparation Cat2 , please replace Cat3 with Cat2 in below line of code.
  final_train_data.rename(columns={'Text':'input','Cat3':'output'},inplace=True)
  final_train_data['text'] = "Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.### Instruction:Classify the text into categories given in output column. Reply with only the words given in output column.### Input:" + final_train_data['input'] + " ### Output:" + final_train_data['output']
  return final_train_data,inference_data
```

```python
final_train_data,inference_data=get_training_data_preparation()
```
